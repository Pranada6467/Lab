```python
import matplotlib.pyplot as plt

# a) Create a figure object
fig = plt.figure()

# b) Add an axis to the figure canvas at [0,0,1,1]
ax = fig.add_axes([0, 0, 1, 1])

# Sample linear data for (x, y)
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]  # This is a linear relationship: y = 2 * x

# c) Plot (x, y) on that axis with a continuous line
ax.plot(x, y, marker='o')  # 'marker="o"' adds markers to highlight data points

# Set labels and title
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Linear Plot')

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure object
fig = plt.figure()

# Add the first axis to the figure at [0,0,1,1]
ax1 = fig.add_axes([0, 0, 1, 1])

# Add the second axis to the figure at [0.2,0.5,.2,.2]
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

# Generate sample data for (x, y)
x = np.linspace(0, 10, 100)  # 100 points between 0 and 10
y = 2 * x  # Linear relationship y = 2 * x

# Plot the data on both axes
ax1.plot(x, y)
ax2.plot(x, y)

# Set labels and title for both axes to distinguish them
ax1.set_xlabel('X-axis Label (ax1)')
ax1.set_ylabel('Y-axis Label (ax1)')
ax1.set_title('Plot on ax1')

ax2.set_xlabel('X-axis Label (ax2)')
ax2.set_ylabel('Y-axis Label (ax2)')
ax2.set_title('Plot on ax2')

# Show the figure
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
# Replace 'company_sales.csv' with the path to your CSV file
df = pd.read_csv('company_sales_data.csv')

# Display the first few rows of the DataFrame to understand its structure
print(df.head())

# Check the columns to ensure 'facewash' is present
print("Columns in the dataset:", df.columns)

# Extract Month and Product Sales columns
months = df['month_number']
products = ['toothpaste', 'bathingsoap', 'facecream', 'facewash', 'shampoo', 'moisturizer']

# Create a figure object
fig = plt.figure()

# Add an axis to the figure canvas
ax = fig.add_axes([0, 0, 1, 1])

# Plot the sales data for each product
for product in products:
    # Check if the column exists in the DataFrame
    if product in df.columns:
        ax.plot(months, df[product], 
                linestyle='-',        # Solid line style
                marker='o',           # Circle marker
                linewidth=2,          # Line width
                label=product.capitalize())  # Label for the legend
    else:
        print(f"Column '{product}' not found in the DataFrame.")

# Set labels and title
ax.set_xlabel('Month Number')
ax.set_ylabel('Units Sold')
ax.set_title('Monthly Units Sold for Each Product')

# Show legend at the lower right location
ax.legend(loc='lower right')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
# Replace 'company_sales.csv' with the path to your CSV file
df = pd.read_csv('company_sales_data.csv')

# Display the first few rows of the DataFrame to understand its structure
print(df.head())

# Assuming 'month_number' is a month index starting from 1 (January) to 12 (December)
# To get the last year's sales data, we'll need to filter for the last 12 months
# Assuming the 'month_number' indicates months in a sequential manner and we're interested in the last year of data

# Identify the last year
# This assumes that 'month_number' is continuous and represents months
last_month = df['month_number'].max()
start_month = last_month - 11  # 12 months in total including the last month

# Filter data for the last year
last_year_data = df[df['month_number'] >= start_month]

# Calculate the total sales for each product for the last year
products = ['toothpaste', 'bathingsoap', 'facecream', 'facewash', 'shampoo', 'moisturizer']
total_sales = {product: last_year_data[product].sum() for product in products}

# Plot the pie chart
fig, ax = plt.subplots()
ax.pie(total_sales.values(), 
       labels=total_sales.keys(), 
       autopct='%1.1f%%',    # Show percentages with one decimal place
       startangle=140,      # Start angle for the pie chart
       colors=plt.cm.Paired(range(len(products))),  # Use a color map
       wedgeprops=dict(width=0.3))  # Make wedges thinner for better display

# Set the title
ax.set_title('Total Sales Distribution for the Last Year')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
# Replace 'company_sales.csv' with the path to your CSV file
df = pd.read_csv('company_sales_data.csv')

# Display the first few rows of the DataFrame to understand its structure
print(df.head())

# Assuming 'month_number' is a month index starting from 1 (January) to 12 (December)
# To get the last year's sales data, we'll need to filter for the last 12 months
# Assuming the 'month_number' indicates months in a sequential manner and we're interested in the last year of data

# Identify the last year
# This assumes that 'month_number' is continuous and represents months
last_month = df['month_number'].max()
start_month = last_month - 11  # 12 months in total including the last month

# Filter data for the last year
last_year_data = df[df['month_number'] >= start_month]

# Calculate the total sales for each product for the last year
products = ['toothpaste', 'bathingsoap', 'facecream', 'facewash', 'shampoo', 'moisturizer']
total_sales = {product: last_year_data[product].sum() for product in products}

# Plot the pie chart
fig, ax = plt.subplots(figsize=(10, 7))  # Increase figure size for better clarity
colors = plt.cm.Paired(range(len(products)))  # Use a color map for distinct colors

# Create the pie chart
wedges, texts, autotexts = ax.pie(total_sales.values(), 
                                  labels=total_sales.keys(), 
                                  autopct='%1.1f%%',    # Show percentages with one decimal place
                                  startangle=140,      # Start angle for the pie chart
                                  colors=colors,       # Apply the color map
                                  wedgeprops=dict(width=0.3))  # Make wedges thinner for better display

# Improve label and percentage visibility
for text in texts:
    text.set_fontsize(12)
    text.set_color('black')

for autotext in autotexts:
    autotext.set_fontsize(12)
    autotext.set_color('white')  # White text on color slices for better contrast

# Set the title
ax.set_title('Total Sales Distribution for the Last Year', fontsize=14)

# Show the plot
plt.show()
```
